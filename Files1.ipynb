```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    traversal = []

    while queue:
        node = queue.popleft()
        traversal.append(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return traversal

# Sample Graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

start_node = 'A'
result = bfs(graph, start_node)
print("BFS Traversal:", result)
```

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    traversal = [start]

    for neighbor in graph[start]:
        if neighbor not in visited:
            traversal += dfs(graph, neighbor, visited)

    return traversal

# Sample Graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

start_node = 'A'
result = dfs(graph, start_node)
print("DFS Traversal:", result)
```

```python
import heapq

def astar(graph, start, goal, heuristic):
    # Priority queue: (f_cost, node)
    open_list = []
    heapq.heappush(open_list, (0, start))

    came_from = {}           # To reconstruct path
    g_cost = {start: 0}     # Cost from start to current node
    f_cost = {start: heuristic[start]}  # g + h

    while open_list:
        _, current = heapq.heappop(open_list)

        # Goal reached
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1], g_cost[goal]

        for neighbor, weight in graph[current].items():
            tentative_g = g_cost[current] + weight

            if neighbor not in g_cost or tentative_g < g_cost[neighbor]:
                g_cost[neighbor] = tentative_g
                f_cost[neighbor] = tentative_g + heuristic[neighbor]
                came_from[neighbor] = current
                heapq.heappush(open_list, (f_cost[neighbor], neighbor))

    return None, float('inf')  # No path found


# Graph (Adjacency list with weights)
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'D': 2, 'E': 5},
    'C': {'A': 4, 'F': 3},
    'D': {'B': 2},
    'E': {'B': 5, 'F': 1},
    'F': {'C': 3, 'E': 1, 'G': 2},
    'G': {'F': 2}
}

# Heuristic values (estimated cost to goal 'G')
heuristic = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 8,
    'E': 3,
    'F': 2,
    'G': 0   # Goal heuristic is always 0
}

start, goal = 'A', 'G'
path, cost = astar(graph, start, goal, heuristic)
print("Path found :", path)
print("Total cost :", cost)
```
